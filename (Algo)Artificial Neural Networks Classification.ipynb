```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

```python
x=df[[ 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked']]
y=df["Survived"]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=True)
```

```python
x_train.shape,  y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
```

# Modeling

**Initializing the Artificial Neural Network**

```python
from keras.models import Sequential
ann=Sequential()
```

**Adding the input layer and the first hidden layer**

```python
from keras.layers import Dense
ann.add(Dense(input_dim = 7,units=6,kernel_initializer = "uniform",activation = "relu"))

# ann.add(Dropout(rate = 0.1))
```

**Adding the second hidden layer**

```python
ann.add(Dense(units=6,kernel_initializer = "uniform",activation="relu"))

# ann.add(Dropout(rate = 0.1)
```

**Adding the output layef**

```python
ann.add(Dense(units=1, kernel_initializer = "uniform",activation="sigmoid"))
```

**Compiling the ANN**

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

**Training the ANN on the Training set**

```python
ann.fit(x_train, y_train, epochs=100, batch_size=32)
```

# Prediction & Evaluating the model

**Predictions**

```python
y_pred = ann.predict(x_train)
y_pred = (y_pred > 0.5)
```

**Evaluation**

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print("Train Accuracy:",accuracy_score(y_train, y_pred))
confusion_matrix(y_train , y_pred)
```

**Cross Validate the model**

```python
def build_cross_classifier():
    classifier = Sequential()
    classifier.add(Dense(input_dim = 7,units=6,kernel_initializer = "uniform",activation = "relu"))
    classifier.add(Dense(units = 6,kernel_initializer = "uniform",activation = "relu"))
    classifier.add(Dense(units = 1,kernel_initializer = "uniform",activation = "sigmoid"))
    classifier.compile(optimizer = "adam",loss = "binary_crossentropy",metrics = ["accuracy"])
    return classifier
```

```python
from scikeras.wrappers import KerasClassifier
classifier = KerasClassifier(build_cross_classifier,batch_size=32,epochs=100)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier,x,y,cv=5)
```
